4.Explain the difference between stemming and lemmatization.
Stemming:
Cuts words to root form without caring about meaning.
Root word may not be correct.
It is fast but less accurate.
"running"->"run"
"studies"->"studi"

Lemmatization:
Converts words to correct base/dictionary form.
Uses grammar and meaning.
It is slow but more accurate.
"studies"->"study"
"better"->"good"

Task 2:Build Advanced Preprocessing Function

```python
import re
def preprocess_text(text):
    #Convert to lowercase
    text = text.lower()

    #Remove numbers
    text = re.sub(r'\d+', '', text)

    #Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    #Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    #Remove punctuation and emojis
    text = re.sub(r'[^\w\s]', '', text)

    #Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    #Remove very short tokens
    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) >= 2 or word in ["no", "not","i"]
    ]

    return " ".join(filtered_words)

# Test data
sentences = [
    "I have 2 dogs",
    "This      is good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "I like this course so much",
    "Visit http://example.com now"
]

# Stress testing
for i, text in enumerate(sentences, 1):
    cleaned = preprocess_text(text)

    print(f"\n{i}:")
    print("Original Text:", text)
    print("Cleaned Sentence:", cleaned)
```

Task 4: Token Analytics

```python
import re
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in ["no", "not"]
    ]

    return filtered_words


sentences = [
    "Get 100% FREE FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever ever... 0/10",
    "Call me at 9876543210",
    "This is THE best best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy happy with this"
]
for i, text in enumerate(sentences, 1):
    tokens = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"\n{i}:")
    print("Tokens:", tokens)
    print("Total tokens:", total_tokens)
    print("Unique tokens:", unique_tokens)
    print("Average token length:", int(round(avg_length, 2)))
```

Task 5: Frequency Analysis

```python
import re
from collections import Counter

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in ["no", "not"]
    ]

    return filtered_words

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

#Combine tokens
all_tokens = []

for text in sentences:
    all_tokens.extend(preprocess_text(text))

# Frequency count
freq = Counter(all_tokens)

# Top 10 most frequent words
top_10 = freq.most_common(10)

# Top 5 least frequent words
least_5 = freq.most_common(5)

print("Top 10 most frequent words:")
for word, count in top_10:
    print(word, ":", count)

print("\nTop 5 least frequent words:")
for word, count in least_5:
    print(word, ":", count)
```

Task 6: Build Full Pipeline

```python
import re
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in ["no", "not"]
    ]

    return filtered_words, " ".join(filtered_words)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned_text = preprocess_text(text)

        all_tokens.extend(tokens) #combine tokens
        clean_sentences.append(cleaned_text)  #store cleaned sentence

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

result = full_pipeline(sentences)
print(result)
```

Task 7: Error Handling

```python
import re
def preprocess_text(text):
    #empty or None input
    if not text or not text.strip():
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    filtered_words = [
        word for word in words
        if len(word) > 2 or word in ["no", "not"]
    ]

    #nothing remains after cleaning
    if not filtered_words:
        return [], ""

    return filtered_words, " ".join(filtered_words)
```
